```python
# This imports a library that will allow us to clear the cell output so we can see only the fresh information for a new room
from tkinter import W
from IPython.display import clear_output
# import the time library to add delays in my code execution
import time

# Global direction variables
#w_north = any 
#s_south = any
#a_east =  any
#d_west =  any



''' -----------------------
      User Movement
------------------------'''
# adding a delay and cleaning up the look of the code
def d(x):
    time.sleep(float(x))

# display the text from the current room you have just entered
def display_content(current_room):
    d(0.05)
    clear_output(wait=True) 
    d(0.05)
    print(f"You have arrived in {current_room[3]}.")
    d(1.5)
    print("Welcome to this room. Story coming soon.") # << pull the room's story from a file named with the rooms title at space [1]
    d(2.5)
    print("How do you proceed?")
    user_input(current_room) 


# user_input is used repeatedly to ask the user which direction they want to move in and when chosen correctly to the move into another room
# current_room = ["rm","Rm0","ri","a large courtyard.","w","Rm1","s","wall","d","wall","a","wall"]

def user_input(current_room):
    user_direction = input("Where would you like to go? Type lowercase 'w' for North, 's' for South, 'a' for West or 'd' for East" )
    # I use all the arrays like key value pairs without acutally making them so in code. Every two entries are related.
    compass = ['w', 'North', 's', 'South', 'd', 'East', 'a', 'West' ] 
    retry = current_room
    #print(user_direction)

    if user_direction == "w" or user_direction == "s" or user_direction == "d" or user_direction == "a":
        clear_output(wait=True)
        # tells you which direction your moving in 
        print(f"Moving {compass[compass.index(user_direction)+1]}")
        d(1.5)
        
        # retrieving the name of the next room from the current rooms array based on the user's input
        next_room = current_room[(current_room.index(user_direction))+1]   

        # if the chosen direction is a wall then tell the user and ask for another entry
        if next_room == 'wall':
            print("You can not go in this direction.")
            user_input(retry)
        else: 
            # uses the string name of the next room to call that rooms function thus moving into the new room
            eval(next_room + "()") 
            
    # can stop the game with this string
    elif user_direction == 'xxx':
        clear_output(wait=True)
        print('manual end')
    
    # catches any characters not recognised as a movement character and asks the user to try again
    else:
        print("Sorry I dont recognise that please try again. Type a lowercase 'w' for North, 's' for South, 'd' for East or 'a' for West" )
        user_input(retry)
     

```
